```python
# Dependencies
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

# Analysis of Price/Rating and Price/Action; applying match function in Excel  and data plotting with Sklearn library

```python
# Load S%P 500 constituent data

kmdata = pd.read_csv('/Users/algend/Desktop/Project_3_Code/StockFinder_Dashboard_Re/static/csv/Financials_2.csv')
kmdata.head(505)
```

# Best(Rating) vs Buy(Action) and/or Better(Rating) vs Sell(Action) and/or Good(Rating) vs Hold(Action) are displaying a numerical value direct match or relative data pertained to the accounts P/R and P/A tested mathematical ratios. 

```python
# Create Array of Samples "X1" for Price/Earnings/Rating(P/R) vs. Price/Earnings/Action(P/A)
X1_df = kmdata.loc[:, kmdata.columns.intersection(['PE/R', 'PE/A'])]
X1 = X1_df.values
#X1
```

```python
plt.scatter(X1[:, 0], X1[:,1])
plt.show()
```

# K-Means clustering is a fast, robust, and simple algorithm that gives reliable results when data sets are distinct or well separated from each other in a linear fashion. 
## https://brilliant.org/wiki/k-means-clustering/

```python
# Use n_clusters=5 as the k value
kmeans = KMeans(n_clusters=4)

# Fit the model to the data
kmeans.fit(X1)
```

```python
# Predict the clusters
predicted_clusters = kmeans.predict(X1)
```

```python
plt.scatter(X1[:, 0], X1[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.xlabel('P/E/Rating')
plt.ylabel('P/E/Action')

plt.show()
```

# Analysis Results:

# The plot is separated into 4 cluster groups from 505 samples:
# Represents a patterned plotted linear correlation between random groups in Price/Earnings/Rating(PE/R) vs. Price/Earnings/Action(PE/A) ratios. Linear correlation assumption is due to price per earnings value present as a part of both the Rating and Action formulas.
